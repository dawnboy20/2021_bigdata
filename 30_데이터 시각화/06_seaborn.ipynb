# 6. seaborn
* matplotlib을 기반으로 다양한 색상테마와 통계용 그래프 기능을 추가한 시각화 라이브러리
* seaborn에서만 제공되는 통계 기반 plot : countplot, boxplot, violinplot, heatmap
* DataFrame을 이용해서 데이터를 시각화하는데 유용함
* https://seaborn.pydata.org/


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
df.describe()
```

```python
df.year.unique()
```

```python
df.year.value_counts()
```

# 6-1. Line Plot
* 선 주위의 파란색은 데이터의 신뢰구간을 표시함

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='year', y='passengers', data=df)
plt.show()
```

# 6-2. Scatter Plot

```python
sns.scatterplot(x='year', y='passengers', data=df)
plt.show()
```

```python
# 겹치지 않게 그려줌 겹치지 않게 그려줌
sns.swarmplot(x='year', y='passengers', data=df)
plt.show()
sns.swarmplot(x='year', y='passengers', data=df)
plt.show()
```

# 6-3. Bar Plot
* x데이터에 대한 y데이터 평균값으로 막대그래프를 그려줌
* 검정색 막대(에러바) 편차

```python
sns.barplot(x='year', y='passengers',data=df)
plt.show()
```

```python
df.loc[df.year == 1960, 'passengers']
```

```python
df.loc[df.year == 1949, 'passengers'].std()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='year', y='passengers', hue='month', data=df)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='day', y='tip', data=tips)
plt.show()
```

## 6-4. Count Plot
* x축에 해당하는 데이터가 몇개 있는지 보여줌

```python
sns.countplot(x='year', data=df)
plt.show()
```

```python
tips.head()
```

```python
sns.countplot(x='day', data=tips)
plt.show()
```

```python
sns.countplot(x='day', hue='time', data=tips)
plt.show()
```

## 6-5. Box Plot (Box-and-Whisker Plot)
* 평균값, 최소값, 최대값, 중앙값, 제 1사분위수, 제3사분위수
* 이상치 (OutLier)를 감지하는데 사용됨

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='year', y='passengers', data=df)
plt.show()
```

```python
df.loc[df.year == 1960].describe()
```
