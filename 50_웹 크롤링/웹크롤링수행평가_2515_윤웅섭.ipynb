### driver를 이용하여 사이트를 열기
  * Selenium으로 연 후 page_source 가져옴
  * 이때, 더보기를 누르지 않으면 데이터의 개수가 한정됨
  * 이에 더보기 버튼 누르는 메서드 구현

```python
driver_path = 'C:\\JupyterLab\\chromedriver.exe'
driver = webdriver.Chrome(driver_path)

driver.get('https://www.sac.or.kr/site/main/show/show_list')
print(driver.title)

time.sleep(1.5)

button = '#pagingGalery > a'
driver.find_element_by_css_selector(button).click()
time.sleep(1.5)
driver.find_element_by_css_selector(button).click()
time.sleep(1.5)
driver.find_element_by_css_selector(button).click()
time.sleep(1.5)
driver.find_element_by_css_selector(button).click()
time.sleep(1.5)
driver.find_element_by_css_selector(button).click()
time.sleep(1.5)
driver.find_element_by_css_selector(button).click()
time.sleep(3)


html = driver.page_source

driver.close()


```

```python
soup = BeautifulSoup(html)
```

### 공연, 일시, 장소 데이터 가져오기

```python
shows = soup.find_all('td', class_ = 'tl')
dates = soup.find_all('td',class_ = 'date')
places = soup.find_all('td', class_='place')
```

### 각 데이터들을 리스트에 요소로 집어넣기

```python
show_title = []

for show in shows:
  show_title.append(show.get_text())
```

```python
date_list = []

for date in dates:
  date_list.append(date.get_text())
```

```python
place_list = []

for place in places:
  place_list.append(place.get_text())
```

### 각 리스트의 요소들을 전체 큰 리스트에 집어넣는 과정

```python
total_list = []
a = len(place_list)

for i in range(a):
  total_list.append([show_title[i], date_list[i], place_list[i]])

total_list
```

### DataFrame화 시키기


```python
columns = ['공연명', '일시', '공연장']
final_df = pd.DataFrame(total_list, columns=columns)
final_df
```

### index 설정해주기

```python
final_df.index += 1
```

### 최종 DataFrame

## xml을 dictionary로

```python
temp  = xmltodict.parse(res.text)
temp
```

## json 형식으로

```python
import json
result = json.loads(json.dumps(temp))
result
```

## 필요한 데이터 따오기

```python
# Airport_list (공항)
# Airline_list (항공사)
# Airport_code (공항코드)
# Exit_list (출구)
# flight_Id (항공코드)
# yoli_list (요일)
items = result['response']['body']['items']['item']
Airport_list = []
Airline_list = []
Maxtem_list = []
Mintem_list = []
wind_list = []
himidity_list = []
flight_Id = []

for item in items:
  Airport_list.append(item['airport'])
  wind_list.append(item['wind'])
  Maxtem_list.append(item['maxtem'])
  Mintem_list.append(item['mintem'])
  Airline_list.append(item['airline'])
  himidity_list.append(item['himidity'])
  flight_Id.append(item['flightId'])
```

## dataframe에 쓸 data 만들기

```python
df_data={
  '공항' : Airport_list,
  '최고온도' : Maxtem_list,
  '최저온도' : Mintem_list,
  '항공사' : Airline_list,
  '습도' : himidity_list,
  '항공코드' : flight_Id
}
```

## 최종 데이터

```python
df = pd.DataFrame(data = df_data)
df
```
