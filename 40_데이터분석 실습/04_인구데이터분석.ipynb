```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('주민등록인구및세대현황_연간.csv', thousands=',')
df.head()
```

```python
df.info()
```

```python
# id_vars : 기준이 되는 컬럼
# value_vars : 컬럼이 녹아서 값과 같이 행으로 내려가는 컬럼
df_2011 = pd.melt(df,id_vars=['행정구역'], value_vars=['2011년_총인구수'], var_name='년도', value_name='인구수')
df_2011.년도 = 2011
df_2011
```

```python
df_2011_m = pd.melt(df,id_vars=['행정구역'], value_vars=['2011년_남자 인구수'], var_name='년도', value_name='남자')
df_2011_m.년도 = 2021
df_2011_m
```

```python
df_2011_f = pd.melt(df,id_vars=['행정구역'], value_vars=['2011년_여자 인구수'], var_name='년도', value_name='여지')
df_2011_f.년도 = 2011
df_2011_f
```

```python
df_2011_all = df.merge(df_2011, df_2011_m, on=['행정구역', '년도'], how='inner')
df_2011_all
```

```python
df_2011_all = df.merge(df_2011_all, df_2011_f, on=['행정구역', '년도'], how='inner')
df_2011_all
```

```python
def population(n):
  # 인구수
  df_temp = pd.melt(df,id_vars=['행정구역'], value_vars=[str(n) + '년_총인구수'], var_name='년도', value_name='인구수')
  df_temp.년도 = n

  # 남자인구수
  df_temp_m = pd.melt(df,id_vars=['행정구역'], value_vars=[str(n) + '년_남자 인구수'], var_name='년도', value_name='남자')
  df_temp_m.년도 = n

  # 여자인구수
  df_temp_f = pd.melt(df,id_vars=['행정구역'], value_vars=[str(n) + '년_여자 인구수'], var_name='년도', value_name='남자')
  df_temp_f.년도 = n

  # merge
  df_temp = pd.merge(df_temp, df_temp_m, on=['행정구역','년도'], how='inner')
  global()['df_{}_all'.format(n)] = pd.merge(df_temp, df_temp_f, on=['행정구역','년도'], how='inner')
  

  # df_2011_all, df_2012_all,.., df_2020_all
```

```python
for i in range(2011, 2021):
  population(i)
```

```python
# df_2011_all = df_2020_all 합치기
df_all = pd.concat([df_2011_all, df_2020_all])
```
