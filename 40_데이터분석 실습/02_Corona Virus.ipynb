```python
df.columns = ['지역','국가','위도','경도','일자','확진자','사망자','회복자','치료중','WHO']
df.head()
```

## 2-5. EDA(탐색적 데이터 분석)

### 1. 대한민국 코로나 데이터 분석

```python
df[df['국가'].str.contains('Korea')]
```

```python
koreaDF = df[df['국가'] == 'South Korea']
koreaDF
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='일자',y='확진자',data=koreaDF)
sns.lineplot(x='일자',y='사망자',data=koreaDF)
sns.lineplot(x='일자',y='회복자',data=koreaDF)
plt.legend(['확진자수','사망자수','회복자수'])
plt.show()
```

### 2. 국가별 코로나 데이터 분석

```python
latestDF = df[df['일자'] == max(df['일자'])]
latestDF.head()
```

```python
sortedDF = latestDF.sort_values(by='확진자', ascending=False)
sortedDF.head(10)
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=45)
sns.barplot(x='국가', y='확진자', data=sortedDF.head(10))
```

```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=45)
sns.barplot(x='국가', y='사망자', data=sortedDF.head(10))
```

```python
globlaDF = df.groupby('일자')[['확진자','사망자','회복자']].sum().reset_index()
globlaDF.head()
```

```python
sns.lineplot(x='일자', y='확진자', data=globlaDF)
```

## 3. folium 사용하기
* 파이썬 기반의 지도 시각화 라이브러리
* http://python-visualization.github.io/folium/

```python
import folium
m = folium.Map()
m
```

```python
# 위도, 경도 데이터 찾기
folium.Map(location=[37.34194526878137, 126.8315083825573], zoom_start=20)
```

```python
m=folium.Map(location=[37.34194526878137, 126.8315083825573], zoom_start=30)
folium.CircleMarker(location=[37.34194526878137, 126.8315083825573], radius=10, color="yellow", fill="Ture", popup="디미고").add_to(m)
m
```

```python
latestDF.head()
```

```python
latestDF.loc[48807, '위도'], latestDF.loc[48807,'경도'], latestDF.loc[48807, '확진자']
```

```python
m = folium.Map(location=[0,0], zoom_start=2)
for n in latestDF.index:
  folium.CircleMarker(location=[latestDF.loc[n, '위도'], latestDF.loc[n,'경도']],
                      radius=latestDF.loc[n,'확진자']/50000,
                      color="red", fill="True",
                      popup=latestDF.loc[n,'국가']+":"+str(latestDF.loc[n,'확진자'])
                      ).add_to(m)
m
```
